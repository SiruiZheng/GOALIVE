```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# deterministic data 
det_GOALIVE = pd.read_csv('deterministic/out_GOALIVE.csv',header=None).to_numpy()
det_single = pd.read_csv('deterministic/out_single.csv',header=None).to_numpy()
det_HER = pd.read_csv('deterministic/out_HER.csv',header=None).to_numpy()
```

```python
det_GOALIVE = np.mean(det_GOALIVE,axis = 1)
det_single = np.mean(det_single,axis = 1)
det_HER = np.mean(det_HER,axis = 1)
```

```python
plt.plot(range(100),det_GOALIVE,label='GOALIVE')
plt.plot(range(100),det_single,label='naive Q learning')
plt.plot(range(100),det_HER,label='HER')
plt.legend()
plt.savefig('deterministic.eps', format='eps')
```
