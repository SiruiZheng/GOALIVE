```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statistics as stats
task_list = [ 'HandReach-v0', 'HandManipulateBlockRotateZ-v0',
            'HandManipulateEggRotate-v0',  'HandManipulatePenRotate-v0']
# task_list = [ 'HandReach-v0','HandManipulateBlockRotateZ-v0']
# 'FetchSlide-v1', 
alg_name = [['her','ddpg'],['goalive_finite','goalive_finite']]
# each task a plot
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs = axs.flatten() 
max_all = []
for kk, task in enumerate(task_list):
    # extract data for plotting
    mid = []
    lb = []
    ub = []
    max_list = []
    for alg in alg_name:
        median_all = []
        data_all = []
        max_sublist = []
        for i in range(5):
            file_name = 'BHER/'+alg[0]+'/experiment/'+alg[1]+'/icml-rebuttal/'+task+'/seed_'+str(i)+'/progress.csv'
            data = list(pd.read_csv(file_name)['test/success_rate'])
            max_sublist.append(max(data))
            #print(len(data))
            #print(file_name)
            median = []
            for j in range(50):
                median.append(stats.median(data[:(i+1)]))
            median_all.append(np.array(median))
            data_all.append(data)
            # print(np.mean(median_all,axis=0).shape)
        max_list.append(np.mean(max_sublist))
        mid.append(np.quantile(data_all,0.5,axis=0))
        lb.append(np.quantile(data_all,0.25,axis=0))
        #print(np.quantile(data_all,0.25,axis=0).shape)
        ub.append(np.quantile(data_all,0.75,axis=0))
    axs[kk].plot(range(50), mid[1], color='blue',label='goalive')
    axs[kk].fill_between(range(50),lb[1],ub[1], color='skyblue')
    axs[kk].plot(range(50), mid[0], color='green',label='ddpg')
    axs[kk].fill_between(range(50),lb[0],ub[0], color='palegreen')
    axs[kk].set_title(task)
    axs[kk].legend()
    max_all.append(max_list)
#plt.show()
plt.savefig('results.jpg', format='jpg')
max_all
```
